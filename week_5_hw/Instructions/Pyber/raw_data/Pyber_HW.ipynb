# Analysis:
#~70% of all rides and total fare value come from Urban riders.
#Rural riders tend to take significantly higher cost, albeit fewer, trips than Urban or Suburban riders.


```python
#import dependencies

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

city_df = pd.read_csv('city_data.csv')
ride_df = pd.read_csv('ride_data.csv')
```

```python
ride_df.head()
```

```python
merged_df = pd.merge(ride_df, city_df, how = "left", on="city")
```

```python
merged_df.head()
```

```python
#filter
merged_df[merged_df["city"] == "Sarabury"]
```

```python
#fulter data into three types of cities
urban = merged_df[merged_df["type"] == "Urban"]
suburban = merged_df[merged_df["type"] == "Suburban"]
rural = merged_df[merged_df["type"] == "Rural"]
```

```python
urban_avg_fare = urban.groupby("city").mean()["fare"]
urban_ride_count = urban.groupby("city").count()["ride_id"]
urban_driver_count = urban.groupby("city").mean()["driver_count"]

suburban_avg_fare = suburban.groupby("city").mean()["fare"]
suburban_ride_count = suburban.groupby("city").count()["ride_id"]
suburban_driver_count = suburban.groupby("city").mean()["driver_count"]

rural_avg_fare = rural.groupby("city").mean()["fare"]
rural_ride_count = rural.groupby("city").count()["ride_id"]
rural_driver_count = rural.groupby("city").mean()["driver_count"]
```

```python
plt.scatter(urban_ride_count, urban_avg_fare, s=10*urban_driver_count, marker='o', edgecolor="black", color = "coral", linewidths=1, alpha=0.8, label="Urban")
plt.scatter(suburban_ride_count, suburban_avg_fare, s=10*suburban_driver_count, marker='o', edgecolor="black", color = "skyblue", linewidths=1, alpha=0.8, label="Suburban")
plt.scatter(rural_ride_count, rural_avg_fare, s=10*rural_driver_count, marker='o', edgecolor="black", color = "gold", linewidths=1, alpha=0.8, label="Rural")


plt.title("insert title")
plt.ylabel("Average Fare")
plt.xlabel("Ride Count")
plt.legend(loc="best")
plt.show()
```

```python
#In addition, you will be expected to produce the following three pie charts:


#% of Total Fares by City Type

#Pie Chart for Fare by city
fare_by_city = 100 * merged_df.groupby(["type"]).sum()["fare"] / merged_df["fare"].sum()

plt.pie(fare_by_city,
        labels=["Rural", "Suburban", "Urban"],
        colors=["gold", "lightskyblue", "lightcoral"],
        explode=[0,0, .09],
        autopct="%1.1f%%", shadow=True, startangle=90)
plt.axis("equal")
#% of Total Rides by City Type
#% of Total Drivers by City Type
plt.show()
```

```python
percent_by_type = merged_df.groupby("type").sum()["fare"] / merged_df["fare"].sum() * 100


plt.pie(percent_by_type, labels = ["Rural", "Suburban", "Urban"], colors=["gold","skyblue","coral"], 
        explode=[0,0,0.1], shadow=True,startangle=150, autopct='%1.1f%%')
plt.title("$ of fares by city type")
plt.axis("equal")
plt.show()
```

```python
rides_by_type = merged_df.groupby("type").count()["ride_id"] / merged_df["ride_id"].count() * 100


plt.pie(rides_by_type, labels = ["Rural", "Suburban", "Urban"], colors=["gold","skyblue","coral"], 
        explode=[0,0,0.1], shadow=True,startangle=150, autopct='%1.1f%%')
plt.title("$ of rides by city type")
plt.axis("equal")
plt.show()
```

```python
drivery_by_type = merged_df.groupby("type").sum()["driver_count"] / merged_df["driver_count"].sum() * 100

#driver_percents = 100 * city_data.groupby(["type"]).sum()["driver_count"] / city_data["driver_count"].sum()
plt.pie(drivery_by_type, labels = ["Rural", "Suburban", "Urban"], colors=["gold","skyblue","coral"], 
        explode=[0,0,0.1], shadow=True,startangle=150, autopct='%1.1f%%')
plt.title("$ of drivers by city type")
plt.axis("equal")
plt.show()
```
