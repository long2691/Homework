```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
!pip install citipy
from citipy import citipy
from random import uniform
```

```python
cities = []

# request up to 500 unique cities
while( len(cities) < 501):
    
    lat = uniform(-90, 90)
    lng = uniform(-180, 180)
    
    city = citipy.nearest_city(lat, lng).city_name
    
    if city not in cities:
        
        cities.append(city)
cities
```

```python
import requests
from config import api_key

# Save config information.
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "metric"
```

```python
#query_url = []
#for city in cities:
    #query_url = f'{url}appid={api_key}&q={city}&units={units}'
   # if query_url not in query_url:
      #  query_url.append(query_url)
lat = []
temp = []
cloud = []
humid = []
windspeed = []
city_not_exist_in_openapi = []
query_url = f"{url}appid={api_key}&units={units}&q="
# = ["Paris", "London", "Oslo", "Beijing"]
# Loop through the list of cities and perform a request for data on each
#Include a print log of each city as it's being processed with the city number, city name, and requested URL.
for city in cities:
    response_url = (query_url + city)
    response = requests.get(response_url).json()
    try:
        lat.append(response['coord']['lat'])
    except KeyError:
        pass
    try:
        cloud.append(response['clouds']['all'])
    except KeyError:
        pass
    try:
        humid.append(response['main']['humidity'])
    except KeyError:
        pass
    try:
        temp.append(response['main']['temp'])
    except KeyError:
        pass
    try:
        windspeed.append(response['wind']['speed'])
    except KeyError:
        pass
    try:
        name = (response['id'])
    except KeyError:
        city_not_exist_in_openapi.append(city)
    print(f"{city}, {name}, {response_url}")
```

```python
weather_dict = {"city": cities,"lat": lat,"temp": temp,"cloudiness": cloud,"humidity": humid,"windspeed": windspeed}
```
